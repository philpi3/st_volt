```python
# Fixing Columns
columns_to_keep = ['Datum', 'Profil', 'Nachricht', 'Anzahl Likes', 'Anzahl Kommentare',
       'Gesamtanzahl Reaktionen, Kommentare & Shares', 'Post-Interaktionsrate','Reichweite pro Post','Interaktionen pro Impression',
       'Profil-ID', 'Beitrag-ID', 'Link']

instagram_1 = instagram[columns_to_keep].copy()
facebook_1 = facebook[columns_to_keep].copy()
twitter_1 = twitter[columns_to_keep].copy()
linkedin_1 = linkedin[columns_to_keep].copy()
tiktok_1 = tiktok24[columns_to_keep].copy()
youtube_1 = youtube[columns_to_keep].copy()

# List of strings to add as columns
platforms = ['Instagram', 'Facebook','Twitter','LinkedIn','TikTok','YouTube']

# List of DataFrames
dfs = [instagram_1, facebook_1, twitter_1, linkedin_1, tiktok_1, youtube_1]

# Add a 'Platform' column with different values to each DataFrame
for i, df in enumerate(dfs):
    platform_value = platforms[i]
    dfs[i]['Platform'] = platform_value

# Resulting modified DataFrames
instagram_2 = dfs[0]
facebook_2 = dfs[1]
twitter_2 = dfs[2]
linkedin_2 = dfs[3]
tiktok_2 = dfs[4]
youtube_2 = dfs[5]
```
